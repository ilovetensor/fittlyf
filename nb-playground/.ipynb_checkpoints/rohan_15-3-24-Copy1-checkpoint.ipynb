### b) Forecasting üë£

### Using ARIMA üå∑

#### Check Stationary
First it is important to check if the series is stationary so that we can implement ARIMA model on the data. 

```python
series = funnel['Actual Productivity']
```

```python
from statsmodels.tsa.stattools import adfuller
from numpy import log
result = adfuller(series)
print('ADF Statistic: ', result[0])
print('p-value: ', result[1])
```

As p-value < 0.05, we go ahead without differentiating, as **the data is stationary**

##### Next, for ARIMA, we have to find values of `p, d, q` for training it.

As the series is Stationary so `d=0`

### Finding order of AR (p)
For value of `p`, we can check the PACF to see how many number of lags have significant lags. 

```python
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
fig, ax = plt.subplots(2, figsize=(10,6))
ax[0].plot(series)

ax[0].set_title('Actual Productivity')
plot_pacf(series, ax=ax[1])

plt.show()
```

##### As PACF lag 1,2 are significant as it is above significant line so we fix **`p = 2`**

### Finding order of MA (q)
For value of `q` we can check ACF to see how many number of lags have significant impact on values.

```python
fig, ax = plt.subplots(2, figsize=(10,6))
ax[0].plot(series)
ax[0].set_title('Actual Productivity')
plot_acf(series, ax=ax[1])    # Calculate ACF
plt.show()
```

##### As ACF has 2 lags significaltly above the line so, set **`q=2`**

### ARIMA Model (2, 0, 2) üì¶

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(series, order=(2,0,2))
model_fit = model.fit()
print(model_fit.summary())
```

Summary of model shows that the coeff have significant values and so the selected values of p,q,d are accurate for this data. 

```python
# Comparing Actual and Predicted values
plt.figure(figsize=(10,3))
preds = model_fit.predict()
plt.plot(preds, label='Predictions')
plt.plot(series, label='Actual')
plt.title('Comparison of Predictions and Actual Values')
plt.xlabel('Period')
plt.ylabel('Actual Productivity')
plt.legend()
plt.grid(True)
plt.show()
```

#### Measuring Accuracy üèπ

```python
def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    mpe = np.mean((forecast - actual)/actual)   # MPE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    return({'Mean Absolute Percentage Error':mape, 'Mean Error':me, 'Mean Absolute Error': mae, 
            'Mean Percentage Error': mpe, 'Root Mean Squared Error':rmse})

forecast_accuracy(preds, series)
```

Around 13% MAPE means the model is about 87% Accurate in predicting next 15 observations

## Rolling Averages üõº

```python
print("Forecaster value :", rolling_avg.iloc[-1])
```

> As can be seen that Rolling Average model also works nicely over the data and captures the seasonal pattern easily for a 7 day span. 

## c) Comparing Model Accuracy üë¨

```python
# Calculating accuracies for both mdoels 
arima = forecast_accuracy(preds, series)
rolling = forecast_accuracy(rolling_avg, series)
compare = pd.DataFrame({'Arima': arima,'Rolling Averages':rolling}).T
compare
```

### 2. AB Testing - Sample Size Calculator üìè

```python
# Given values
MDE = 0.03  # Minimum Detectable Effect (3%)
alpha = 0.05  # Significance level (95%)
power = 0.8  # Statistical power (80%)

# p = ?     -> Baseline Conversion Rate
```

### Calculating Baseline Conversion Rate for data 
$ Conversion = Clicks / Visitors $

```python
abtest['Conversion'] = abtest['Clicks']/abtest['Visitors']
abtest.drop(abtest.index[abtest.Visitors==0], inplace=True)
baseline_conversion = abtest['Conversion'].mean()
print('Baseline Conversion Rate : ', baseline_conversion)
```

## Sample Size Calculator üê¨

$$ size = \frac{( {z_{\alpha/2}} + {z_{\beta}} )^2 \cdot p \cdot (1 - p)} {MDE^2}$$



```python
def calculate_sample_size(alpha, beta, baseline_conversion, mde):
    # Standard normal critical values at alpha/2 and 1 - beta
    z_alpha_2 = norm.ppf(1 - alpha / 2)
    z_beta = np.abs(norm.ppf(1 - beta))
    
    # Calculate sample size
    sample_size = ((z_alpha_2 + z_beta)**2 * baseline_conversion * (1 - baseline_conversion) )/ (mde**2)
    
    return int(round(sample_size)) *  2
```

```python
sample_size_per_group = calculate_sample_size(alpha, power, baseline_conversion, MDE)

print("Required sample size per group:", sample_size_per_group)
```

- hence with the formula we have calculated the required sample size
- So, it is more than sufficient to proove the hypothesis

### 3. Hypothesis Test

```python
def perform_ab_test(control_visitors, control_conversions, treatment_visitors, treatment_conversions, confidence_level):
    
    # Calculate conversion rates
    control_rate = control_conversions / control_visitors
    treatment_rate = treatment_conversions / treatment_visitors
    
    # Calculate pooled standard error 
    pooled_se = np.sqrt(control_rate * (1 - control_rate) / control_visitors + treatment_rate * (1 - treatment_rate) / treatment_visitors)
    
    # Calculate Z-score
    z_score = (treatment_rate - control_rate) / pooled_se
    
    # Calculate p-value using the standard normal distribution 
    p_value = 2 * (1 - norm.cdf(abs(z_score)))
    
    # Compare p-value with significance level 
    significance_level = 1 - confidence_level / 100
    
    if p_value < significance_level:
        if z_score > 0:
            return "Experiment Group is Better"
        else:
            return "Control Group is Better"
    else:
        return "Indeterminate"
```

```python
control_visitors = abtest[abtest['Group'] == 'Control']['Visitors'].sum()
control_conversions = abtest[abtest['Group'] == 'Control']['Clicks'].sum()
treatment_visitors = abtest[abtest['Group'] == 'Experiment']['Visitors'].sum()
treatment_conversions = abtest[abtest['Group'] == 'Experiment']['Clicks'].sum()

result = perform_ab_test(control_visitors, control_conversions, treatment_visitors, treatment_conversions, confidence_level=98)
print("AB Test Result:", result)
```
